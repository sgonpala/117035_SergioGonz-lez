```python
'''Leer un número entero y determine el número de cincos que están 
consecutivos en un número. Por ejemplo, en 5512355551 hay 6 cincos 
consecutivos.'''


while True:                         # Solicitud de datos y catch de errores
    try:
        x = int(input("Ingrese un número entero: "))
        x = abs(x)
        break
    except ValueError:
        print("POR FAVOR INGRESE UN NÚMERO ENTERO.")

cinco_rep = 0                  # Definición de las variables necesarias

while x > 0:

    a = x % 10
    b = (x // 10) % 10

    if a == b == 5:           # Análisis de los dígitos y si son consecuttivos
        if (x // 100) % 10 == 5:
            cinco_rep += 1
            x // 10
        else:
           cinco_rep += 2

    x = x // 10

if cinco_rep:                  # Impresión de resultados
    print("\nEl número 5 está", cinco_rep, "veces consecutivas en el número.")

else:
    print("\nEn el número no hay 5 consecutivos")
```

```python
'''Recibe un número entero y define si la suma de sus dígitos pertenece a la \
serie fibonacci'''


def serie_fibonacci(x):

    '''Regresa verdadero o falso si el número ingresado pertenece o no a la \
    serie fibonacci'''

    a = 0
    b = 1
    s = 0

    while s < x:
        s = a + b
        a = b
        b = s
        if s == x:
            return True
    return False


while True:
    try:
        num = int(input("Por favor ingrese un número entero: "))
        if num < 0:
            num = abs(num)
            break                      # Solicitud de datos y catch de errores
        else:
           num=num
           break
    except ValueError:
        print("POR FAVOR INGRESE UN NÚMERO ENTERO: ")

suma_dig = 0

while num > 0:
    digito = num % 10
    num = num // 10                       # Suma de los dígitos del número
    suma_dig += digito

if serie_fibonacci(suma_dig):
    print("La suma de los dígitos del número ingresado es", suma_dig,
          "y sí está en la serie Fibonacci")

else:
    print("La suma de los dígitos del número es", suma_dig,
          "y no está en la serie Fibonacci")
```

```python
'''Programa que lee las coordenadas (x1,y1,r1) y (x2,y2,r2) que corresponden \
al centro y al radio de dos círculos. Lee un punto de coordenadas (a,b).
Determina si (a,b) está contenido dentro del círculo 1, dentro del círculo 2 o\
 dentro de ambos círculos.'''


def dentro_del_circulo(xc=0.0, yc=0.0, r=1.0, xi=0.0, yi=0.0):
    '''
    Determina si un punto dado con coordenadas xi,yi está dentro de la \
    circunferencia dada por el centro xc,yc y radio r

    Coordenada en x del centro :param xc:
    Coordenada en y del centro :param yc:
    Radio de la circunferencia :param r:
    Coordenada x del punto para analizar :param xi:
    Coordenada y del punto para analizar :param yi:
    True or False :return:
    '''

    return ((xc - r) <= xi <= (xc + r)) and ((yc - r) <= yi <= (yc + r))


while True:
    try:
        x1 = float(input("Coordenada x del centro del círculo 1: "))
        y1 = float(input("Coordenada y del centro del círculo 1: "))
        r1 = float(input("El radio del círculo 1: "   ))               
        x2 = float(input("Coordenada x del dentro del círculo 2: "))
        y2 = float(input("Coordenada y del centro del círculo 2: "))
        r2 = float(input("El radio del círculo 2: "))
        a  = float(input("Coordenada x del punto: "))  #Solicitud de datos y 
        b  = float(input("Coordenada y del punto: "))  #catch de errores
        break

    except ValueError:
        print("POR FAVOR INGRESE UN NÃMERO.")


# Impresión de resultados de acuerdo a si el punto está dentro de ambos 
#círculos, de uno de los dos o en ninguno de ellos.


if dentro_del_circulo(x1, y1, r1, a, b) and dentro_del_circulo(x2, y2, r2,\
                     a, b):
    print("El punto ({0} , {1}) se encuentra dentro del círculo 1 y 2."\
          .format(a, b))

elif dentro_del_circulo(x1, y1, r1, a, b):
    print("El punto ({0} , {1}) se encuentra dentro del círculo 1."\
          .format(a, b))

elif dentro_del_circulo(x2, y2, r2, a, b):
    print("El punto ({0} , {1}) se encuentra dentro del círculo 2."\
          .format(a, b))

else:
    print("El punto ({0} , {1}) no se encuentra dentro de ningún círculo."\
          .format(a, b))
```

```python
'''Recibe una cadena de texto y la imprime en mayúscula'''


cad = input("Ingrese una cadena de texto: ")  # Se solicita la cadena de texto
cad_copia = ""

for i in cad:
    if not i.isalpha():
        cad_copia += i
    if i in "AEIOUÁÉÍÓÚQWERTYÜPSDFGHJKLÑZXCVBNM":
        cad_copia += i
    elif 97 <= ord(i) <= 122:
        cad_copia += chr(ord(i) - 32)    
    elif i == "á":                     
        cad_copia += "Á"         # Análisis de la oración letra por letra     
    elif i == "é":               # y su conversión a mayúscula, adición 
        cad_copia += "É"         #a la nueva cadena
    elif i == "í":
        cad_copia += "Í"
    elif i == "ó":
        cad_copia += "Ó"
    elif i == "ú":
        cad_copia += "Ú"
    elif i == "ü":
        cad_copia += "Ü"
    elif i == "ñ":
        cad_copia += "Ñ"

print("la cadena de texto ingresada pero en mayúscula es: ",{cad_copia})
```

```python
"""Lee una lista de números ordenados de forma ascendente, verifica que esté 
ordenada, lee otro número y lo ubica en la posicicón adecuada"""


class A(Exception):                               # Definición de la excepción
    """La lista no se encuentra ordenada."""
    pass


while True:
    try:
        lista1 = eval(input("Ingrese una lista de números ordenada: "))  
                                                     # Solicitud de la lista
        for i in range(1, len(lista1) - 1):
            if lista1[i - 1] <= lista1[i] <= lista1[i + 1]:# Análisis de orden
                pass
            else:
                raise A
        else:
            break

    except A:
        print("LA LISTA NO ESTÁ ORDENADA.")

while True:
    try:                              # Solicitud del elemento para ordenar
        ele = float(input("Ingrese un número para insertar en la lista: "))
        print()
        break
    except ValueError:
        print("INGRESE UN NÚMERO ADECUADO.")


for i in range(len(lista1)):       # Insertar el elemento en el orden adecuado
    if lista1[i] > ele:
        lista1.insert(i, ele)
        break
else:
    lista1.append(ele)

print("La lista con el elemento adicional en la posición correcta es:", lista1)



```
